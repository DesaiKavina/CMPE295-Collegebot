```python
import requests    
import re    
from urllib.parse import urlparse    

class PyCrawler(object):    
    def __init__(self, starting_url):    
        self.starting_url = starting_url    
        self.visited = set()    

    def get_html(self, url):    
        try:    
            html = requests.get(url)    
        except Exception as e:    
            print(e)    
            return ""    
        return html.content.decode('latin-1')    

    def crawl(self, url):
        html = self.get_html(url) 
        print(html)
        questions = re.findall("<p><strong>Q\.(.*?)</strong><br>",html,flags=re.S)    
        answers = re.findall("</strong><br>(.*?)</p>",html, flags=re.S)
        
        print(len(questions))         
        print(len(answers)) 
        result = []
        links = []
        for ans in answers:
            linkans = re.findall('''<a\s+(?:[^>]*?\s+)?href="([^"]*)"''', ans) 
            ans = ans.replace("A. ","")            
            ans = ans.replace("<span>","")
            ans = ans.replace("</span>","")           
            ans = ans.replace("<strong>"," ")
            ans = ans.replace("</strong>","")
            ans = ans.replace("<br>","")
            ans = ans.replace("&nbsp;","")            
            ans = ans.replace("</a>","")
#             print(linkans)
            ans = re.sub("(<a href=.*?>)", "",ans) 
            result.append(ans)
            temp = " "
            temp = temp.join(linkans)
            links.append(temp)
        


        import csv

        with open('some4.csv', 'w') as f:
            writer = csv.writer(f)
            writer.writerows(zip(questions, result, links))
#         print(answers)    
  

    def start(self):
        self.crawl(self.starting_url)  
#         import pandas as pd 
#         data = pd.read_csv("sheet6.csv") 
#         ulit = data.values.tolist()
#         for u in ulit:
# #             print(u[0])
#             self.crawl(u[0])
        
        

if __name__ == "__main__":    
    
    crawler = PyCrawler("http://www.sjsu.edu/gape/faq/") 
    crawler.start()
```
